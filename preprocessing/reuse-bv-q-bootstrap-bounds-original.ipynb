```r
## 2019-01-07 -- Chase DL Errors
df = read.csv('C:/Users/jchang/Desktop/Projects/Funnel Anomaly Detection/data/dataexportdl_2019-1-25_1505.csv')
```

```r
df$dl = ifelse(is.na(df$dl) == TRUE, 0, df$dl)
df$q = ifelse(is.na(df$q) == TRUE, 0, df$q)
df$hist_dl = ifelse(is.na(df$hist_dl) == TRUE, 0, df$hist_dl)

currentdate = as.Date(df$timestart[which(is.na(df$timestart)==FALSE)], format = "%Y-%m-%d %H:%M:%S")
currentdate = max(currentdate[which(is.na(currentdate)==FALSE)])
currentday = mean(df$dayofweek[which(df$dl != 0)])
# currenthours = unique(df$hourofday[which(df$dl != 0)])
currenthours = seq(0,max(df$hourofday))


histsums_dl = aggregate(x = df$dl, FUN = sum, by = list(df$dayofweek, df$hourofday))
colnames(histsums_dl) = c("dayofweek", "hourofday", "sum_dl")

histsums_q = aggregate(x = df$q, FUN = sum, by = list(df$dayofweek, df$hourofday))
colnames(histsums_q) = c("dayofweek", "hourofday", "sum_q")

point_est = data.frame(histsums_q$dayofweek, histsums_q$hourofday, histsums_dl$sum_dl/histsums_q$sum_q)
colnames(point_est) = c("dayofweek", "hourofday", "point_est")
point_est = point_est[which(point_est$dayofweek == currentday & point_est$hourofday %in% currenthours),]


histsums_histdl = aggregate(x = df$hist_dl, FUN = sum, by = list(df$dayofweek, df$hourofday))
colnames(histsums_histdl) = c("dayofweek", "hourofday", "sum_dl")

histsums_histq = aggregate(x = df$hist_q, FUN = sum, by = list(df$dayofweek, df$hourofday))
colnames(histsums_histq) = c("dayofweek", "hourofday", "sum_q")

hist_r = data.frame(histsums_histq$dayofweek, histsums_histq$hourofday, histsums_histdl$sum_dl/histsums_histq$sum_q)
colnames(hist_r) = c("dayofweek", "hourofday", "r")

hist_n = data.frame(histsums_histq$dayofweek, histsums_histq$hourofday, histsums_histq$sum_q)
colnames(hist_n) = c("dayofweek", "hourofday", "n")

histsd_histq = aggregate(x = df$hist_q, FUN = sd, by = list(df$dayofweek, df$hourofday))
colnames(histsd_histq) = c("dayofweek", "hourofday", "sd_q")

histsd_histdl = aggregate(x = df$hist_dl, FUN = sd, by = list(df$dayofweek, df$hourofday))
colnames(histsd_histdl) = c("dayofweek", "hourofday", "sd_dl")

histavg_histq = aggregate(x = df$hist_q, FUN = mean, by = list(df$dayofweek, df$hourofday))
colnames(histavg_histq) = c("dayofweek", "hourofday", "mean_q")




corr_mat = matrix(nrow = length(currenthours), ncol = 3)
for(i in 1:length(currenthours)){
  q = df$hist_q[which(df$dayofweek == currentday & df$hourofday == i-1)]
  dl = df$hist_dl[which(df$dayofweek == currentday & df$hourofday == i-1)]
  corr_mat[i,3] = cor(q,dl)
  corr_mat[i,1] = currentday
  corr_mat[i,2] = i-1
}

hist_corr = data.frame(corr_mat)
colnames(hist_corr) = c("dayofweek", "hourofday", "corr")


hist_n = hist_n[which(hist_n$dayofweek == currentday & hist_n$hourofday %in% currenthours),]
hist_r = hist_r[which(hist_r$dayofweek == currentday & hist_r$hourofday %in% currenthours),]
histsd_histdl = histsd_histdl[which(histsd_histdl$dayofweek == currentday & histsd_histdl$hourofday %in% currenthours),]
histsd_histq = histsd_histq[which(histsd_histq$dayofweek == currentday & histsd_histq$hourofday %in% currenthours),]
histavg_histq = histavg_histq[which(histavg_histq$dayofweek == currentday & histavg_histq$hourofday %in% currenthours),]
hist_corr = hist_corr[which(hist_corr$dayofweek == currentday & hist_corr$hourofday %in% currenthours),]
se = sqrt((hist_r[,3]^2*histsd_histq[,3]^2 + histsd_histdl[,3]^2 - 2*hist_r[,3]*hist_corr[,3]*histsd_histq[,3]*histsd_histdl[,3])/(hist_n[,3]*histavg_histq[,3]^2))


z_upper = vector(length=length(currenthours))
z_lower = vector(length=length(currenthours))
ci_information = data.frame(hist_n$dayofweek, hist_n$hourofday, hist_r$r, se, z_upper, z_lower)


B=5000
bootstrap_z = vector(length = B)
alpha = 1/20

start_time = Sys.time()
for(i in 1:length(currenthours)){
  for(j in 1:B){
    df_temp = df[which(df$dayofweek == currentday & df$hourofday == i-1),]
    df_temp$hist_dl = ifelse(is.na(df_temp$hist_dl) == TRUE, 0, df_temp$hist_dl)
    df_temp$hist_q = ifelse(is.na(df_temp$hist_q) == TRUE, 0, df_temp$hist_q)
    
    index = seq(1, nrow(df_temp))
    resample_index = sample(x = index, size = length(index), replace = TRUE)
    
    resample_mean = sum(df_temp$hist_dl[resample_index])/sum(df_temp$hist_q[resample_index])
    mean = ci_information$hist_r.r[i]
    
    n = sum(df_temp$hist_q[resample_index])
    r = resample_mean
    sx = sd(df_temp$hist_q[resample_index])
    sy = sd(df_temp$hist_dl[resample_index])
    mx = mean(df_temp$hist_q[resample_index])
    corr = cor(df_temp$hist_dl[resample_index], df_temp$hist_q[resample_index])
    
    resample_se = sqrt((r^2*sx^2 + sy^2 - 2*r*corr*sx*sy)/(n*mx^2))
    
    bootstrap_z[j] = (resample_mean - mean)/resample_se
  }
  bootstrap_z = sort(bootstrap_z)
  ci_information$z_lower[i] = bootstrap_z[alpha/2*B]
  ci_information$z_upper[i] = bootstrap_z[(1-alpha/2)*B]
}
elapsed_time = Sys.time() - start_time
elapsed_time

lower = ci_information$hist_r.r - ci_information$z_upper*ci_information$se
upper = ci_information$hist_r.r - ci_information$z_lower*ci_information$se

ci_information = data.frame(ci_information
                            , currentdate
                            , point_est
                            , lower
                            , upper)
colnames(ci_information) = c("dayofweek", "hourofday", "hist_mu", "hist_se", "z_upper", "z_lower", "date", "dow", "hod", "point_est", "lower", "upper")
ci_information_legacy = ci_information
ci_information_legacy
```
